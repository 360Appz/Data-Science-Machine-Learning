```python
import pandas as pd
from neuralprophet import NeuralProphet
from matplotlib import pyplot as pt


#Have to import because of some Tensorflow related issues
import os
os.environ['KMP_DUPLICATE_LIB_OK']='True'
```

```python
df = pd.read_csv('weatherAUS.csv')
df.head()
```

```python
df.Location.unique()
```

```python
df.columns #View columns
```

```python
df.dtypes
```

```python
melb = df[df['Location']=='Melbourne'] #Data preprocessing, filter out Melbourne
melb['Date'] = pd.to_datetime(melb['Date']) #Convert date object to a proper date for melbourne
melb.head()

```

```python
melb.dtypes #Check to confirm that date type has been changed
```

```python
pt.plot(melb['Date'], melb['Temp3pm'])
pt.show()
```

```python
melb['Year'] = melb['Date'].apply(lambda x: x.year) #Lambda used to extract the year
melb = melb[melb['Year']<=2015] #Removes missing data chunk from 2015-2016
pt.plot(melb['Date'], melb['Temp3pm'])
pt.show()
```

```python
melb.tail()
```

```python
data = melb[['Date', 'Temp3pm']] #We just need the date and temperature
data.dropna(inplace=True) # Filter out other columns
data.columns = ['ds', 'y'] 
data.head()
```

```python
m = NeuralProphet()
model = m.fit(data, freq='D', epochs=1000) #Neural network fitted with data on daily frequency, epoch shows the amount of time the algorithm will workthrough the dataset
```

```python
future = m.make_future_dataframe(data, periods=900) #Forecast for 900 period
forecast = m.predict(future)
forecast.head()
```

```python
forecast.tail()
```

```python
plot1 = m.plot(forecast) #Temperature highest during year end
```

```python
plt2 = m.plot_components(forecast) #Break down temperatures by seasons
```
